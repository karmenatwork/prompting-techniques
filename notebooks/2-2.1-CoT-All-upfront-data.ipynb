# Chain of Thought

“Think step by step” with all the context you provide is powerful, but it’s largely static in this sense:

The model can only reason over the information that’s already in the prompt/context window.

It won’t go fetch new facts, run real tools, observe the world, check a database, call an API, or update itself mid-response unless your system is explicitly set up to do that (e.g., tool use / function calling / agent loop).

So the “chain” is a one-shot derivation from a fixed snapshot of context.

```python
from openai import OpenAI
import os
from utils import display_responses, get_completion 

from dotenv import load_dotenv, find_dotenv
_ = load_dotenv(find_dotenv()) # read local .env file


client = OpenAI(
    api_key=os.getenv("OPENAI_API_KEY")
    )
MODEL = "gpt-4o-mini"
```

```python
def chat(prompt: str, system: str = "You are a helpful assistant"):
    try: 
        response = client.chat.completions.create(
            model="gpt-4o-mini",
            messages=[
                {"role": "system", "content": system},
                {"role": "user", "content": prompt},
            ],
            temperature=0.7,
        )
        content = response.choices[0].message.content
        # print(content)
        return content
    except Exception as e:
        print(f"An error ocurred:{e}")
```

```python
system_prompt ="""You are an expert sports analyst and event planner.

Think step by step and connect the dots carefully.

The FIFA World Cup 2026 will be hosted by three countries: 
the United States, Canada, and Mexico.

Based on this information:
1. Consider geography and travel distances
2. Consider infrastructure and stadium availability
3. Consider coordination challenges across borders

Explain step by step why hosting the tournament across three countries
is both an opportunity and a challenge.

Then provide a concise final summary.
"""
user_prompt = """The FIFA World Cup 2026 will be hosted by three countries:
the United States, Canada, and Mexico.

Based on this information, analyze why hosting the tournament across
three countries is both an opportunity and a challenge.

In your analysis, consider:
1. Geography and travel distances
2. Infrastructure and stadium availability
3. Cross-border coordination and logistics

Provide a concise explanation followed by a short final summary."""
```

```python
response = chat(user_prompt, system_prompt)
display_responses({
    "system_prompt": system_prompt,
    "user_prompt": user_prompt,
    "response": response
})
```
